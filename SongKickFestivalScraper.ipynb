```python
import pandas as pd
import string
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup as bs
import pickle
%matplotlib inline
```

```python
page_prefix = 'https://www.songkick.com/festivals/1234-bonnaroo-music/id/18350264-bonnaroo-music-festival-2014'
page = requests.get(page_prefix)
page
```

```python
soup = bs(page.text)
```

```python
link_dict = {}
fest_list = soup.find('ul', class_="festival")
for tag in fest_list.find_all('li'):
    link_dict[tag.find('a').get('href')] = tag.text
link_dict
```

```python
with open('MillionSongCode/artists.pkl', 'r') as f:
    million_artists = pickle.load(f)
```

```python
million_artists
```

```python
overlappers = []
for artist in million_artists:
    if artist in link_dict.values():
        overlappers.append(artist)
overlappers
```
